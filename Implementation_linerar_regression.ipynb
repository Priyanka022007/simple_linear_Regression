```python
#Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#IMport dataset

ds=pd.read_csv('Salary_Data.csv')
ds
```

```python
ds.columns
```

```python
#Divide dataset as independent(input) and dependent(output)

x=ds[['YearsExperience']]
y=ds[['Salary']]
```

```python
x
```

```python
y
```

```python
#plot graph -for Visualiztion (How it work on it) 
plt.scatter(x,y) #Scatter plot used because we use 2 variables
plt.show()
```

```python
#spliting dataset as Training and Testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train  #We can use trainning dataset for model fitting
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Introduce Model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
#Model will be define now
```

```python
#Predict  -salary

result=reg.predict([[5]])
result
#When which years of experience on that it gaive us their salary
```

```python
#Predict

result=reg.predict(x_test)
result
```

```python
y_test
```

```python
# visualiztion plot -To check get outliers or not

plt.scatter(x,y)
result=reg.predict(x)
plt.plot(x,result,color='pink') #color of line
plt.show()
```
